**Data augmentation** is a set of techniques to artificially increase the amount of data by generating new data points from existing data. This includes making small changes to data or using deep learning models to generate new data points.
**Image augmentation** is a technique of altering the existing data to create some more data for the model training process. In other words, it is the process of artificially expanding the available dataset for training a deep learning model.

```python
img_gen = ImageDataGenerator()

img = img_gen.flow(image, batch_size=1)
fig,ax = plt.subplots(nrows = 1,ncols = 3,figsize = (15,15))

# Genreate batch of images
for i in range(3):
  #convert to unsigned integers
  shihtzu = next(img)[0].astype('uint8')
  #plot image
  ax[i].imshow(shihtzu)
  ax[i].axis('on')
```

**Random Rotation**  rotate the image clockwise or anti-cloackwise randomly between the specified range.

```python
# ImageDataGenerator Instance creation
img_gen = ImageDataGenerator(rotation_range=110)

img = img_gen.flow(image, batch_size=1)
fig,ax = plt.subplots(nrows = 1,ncols = 4,figsize = (15,15))

# Genreate batch of images
for i in range(4):
  #convert to unsigned integers
  shihtzu = next(img)[0].astype('uint8')
  #plot image
  ax[i].imshow(shihtzu)
  ax[i].axis('on')
```

**Flip** - Flip the image along the vertical or the horizontal axis.

```python
# Horizontal flip, vertical flip
# ImageDataGenerator Instance creation
img_gen = ImageDataGenerator(horizontal_flip=True, vertical_flip=True)

img = img_gen.flow(image, batch_size=1)
fig,ax = plt.subplots(nrows = 1,ncols = 3 ,figsize = (15,15))

# Genreate batch of images
for i in range(3):
  #convert to unsigned integers
  shihtzu = next(img)[0].astype('uint8')
  #plot image
  ax[i].imshow(shihtzu)
  ax[i].axis('on')
```

**Random shift** -  The height_shift_range argument is used for a vertical shift of the image and width_shift_range is used for a horizontal shift of image

```python
img_gen = ImageDataGenerator(width_shift_range=0.3, height_shift_range=0.2,fill_mode = 'nearest') 

# ImageDataGenerator instance creation
aug_iter = img_gen.flow(image,batch_size = 1)

fig,ax = plt.subplots(nrows = 1,ncols = 3,figsize = (15,15))

# generate batch of images
for i in range(3):
  #convert to unsigned integers
  shihtzu = next(aug_iter)[0].astype('uint8')
  #plot image
  ax[i].imshow(shihtzu)
  ax[i].axis('on')
```

**Zoom** used to randomly zoom in and out of the image.  If the value is float type, then the zoom will be done in the range [1 - zoom_range, 1+zoom_range].Any value smaller than 1 zoom in on the image. Any value greater than 1 zoom out on the image.

```python
img_gen = ImageDataGenerator(zoom_range=0.6)

img = img_gen.flow(image, batch_size=1)
fig,ax = plt.subplots(nrows = 1,ncols = 3 ,figsize = (15,15))

# Genreate batch of images
for i in range(3):
  #convert to unsigned integers
  shihtzu = next(img)[0].astype('uint8')
  #plot image
  ax[i].imshow(shihtzu)
  ax[i].axis('on')
```

**Brightness range** - Change brightness of the image if object is not under perfect lighting condition. Values less than 1 darkens the image and values above 1 brightens the image.

```python
img_gen = ImageDataGenerator(brightness_range=[0.5,2.5])

img = img_gen.flow(image, batch_size=1)
fig,ax = plt.subplots(nrows = 1,ncols = 3 ,figsize = (15,15))

# Genreate batch of images
for i in range(3):
  #convert to unsigned integers
  shihtzu = next(img)[0].astype('uint8')
  #plot image
  ax[i].imshow(shihtzu)
  ax[i].axis('on')
```

**Shear range** - Shear means image will be distorted along an axis inorder to create or rectify the perception angles. It's usually used to augment images so that computers can see how humans see things from different angles.

```python
img_gen = ImageDataGenerator(shear_range=60)

img = img_gen.flow(image, batch_size=1)
fig,ax = plt.subplots(nrows = 1,ncols = 3 ,figsize = (15,15))

# Genreate batch of images
for i in range(3):
  #convert to unsigned integers
  shihtzu = next(img)[0].astype('uint8')
  #plot image
  ax[i].imshow(shihtzu)
  ax[i].axis('on')
```

**Channel shift range** - A random channel shift will be done based on the specified fixed value of the RGB Color Channels of the input image

```python
img_gen = ImageDataGenerator(channel_shift_range=100)

img = img_gen.flow(image, batch_size=1)
fig,ax = plt.subplots(nrows = 1,ncols = 3 ,figsize = (15,15))

# Genreate batch of images
for i in range(3):
  #convert to unsigned integers
  shihtzu = next(img)[0].astype('uint8')
  #plot image
  ax[i].imshow(shihtzu)
  ax[i].axis('on')
```

**Rescaling** - Original image consists of RGB coefficients in 0-255 range, but such values would be too high for our model to process (given a typical learning rate), so we target values between 0 and 1 instead of by scaling with a 1/255.

```python
img_gen = ImageDataGenerator(rescale=1./255)

img = img_gen.flow(image, batch_size=1)
fig,ax = plt.subplots(nrows = 1,ncols = 3 ,figsize = (15,15))

# Genreate batch of images
for i in range(3):
  #convert to unsigned integers
  shihtzu = next(img)[0].astype('float32')
  #plot image
  ax[i].imshow(shihtzu)
  ax[i].axis('on')
```
